## DATA CLEANING FOR IMBALANCED DATA ( UNDERSAMPLING)

# TOMEK'S LINK

```python
%pip install imbalanced-learn
```

```python
from imblearn.under_sampling import TomekLinks

# Separate features and target
X = parkinsons_dataset.drop(columns=['name', 'status'], axis=1)
y = parkinsons_dataset['status']

# Apply Tomek's Link
tl = TomekLinks()
X_res, y_res = tl.fit_resample(X, y)

# Check the shape of the resampled data
print("Original dataset shape:", X.shape, y.shape)
print("Resampled dataset shape:", X_res.shape, y_res.shape)
```

```python
# Split the resampled data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
# Train the model
model = svm.SVC(kernel='linear')
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(Y_test, y_pred))
```

```python
# print results
print("confusion_matrix : ")
print(conf_matrix)
```

```python
print("\nclassification_report : ")
print(classification_report(Y_test, y_pred))
```

```python

from sklearn.metrics import f1_score

f1 = f1_score(Y_test, y_pred)

print("F1 Score: ", f1)
```

```python
import pickle
filename = 'parkinsons_model.sav'
pickle.dump(model, open(filename, 'wb')) ##wb = write byte
```
